```python
import pandas as pd
import re

#df = pd.read_csv("merged_unique.csv")

def clean_bio(text):
    if pd.isna(text):
        return ""

    # Remove page marker: [ ص: 477 ]
    text = re.sub(r"\[\s*ص:\s*\d+\s*\]", "", text)

    # Remove narrator grade ( ع ) when isolated
    text = re.sub(r"\(\s*ع\s*\)", "", text)

    # Remove double newlines
    text = text.replace("\n\n", " ").replace("\n", " ")

    # Remove multiple spaces
    text = re.sub(r"\s+", " ", text)

    # Remove starting/ending spaces
    text = text.strip()

    return text

df["bio_clean"] = df["bio"].apply(clean_bio)

df.to_csv("merged_cleaned.csv", index=False, encoding="utf-8")

print("Cleaning complete!")
print("Rows:", len(df))
print(df[["bio", "bio_clean"]].head())
```

```python
import re
import json
from typing import Dict, List, Optional

class NarratorBiographyExtractor:
    """
    Extracts structured metadata from Arabic biographical texts of hadith narrators.
    Designed for texts from classical Islamic biographical works like Siyar A'lam al-Nubala.
    """

    def __init__(self):
        self.hijri_to_gregorian_offset = 622  # Approximate conversion

    def extract_narrator_metadata(self, arabic_text: str, narrator_id: str = "N001") -> Dict:
        """
        Main extraction method that processes Arabic biographical text.

        Args:
            arabic_text: The Arabic biographical text
            narrator_id: Unique identifier for the narrator

        Returns:
            Dictionary containing structured narrator metadata
        """
        metadata = {
            "narrator_id": narrator_id,
            "full_name": self._extract_full_name(arabic_text),
            "aliases": self._extract_aliases(arabic_text),
            "birth_year": self._extract_birth_year(arabic_text),
            "death_year": self._extract_death_year(arabic_text),
            "birthplace": self._extract_birthplace(arabic_text),
            "primary_locations": self._extract_locations(arabic_text),
            "era": self._determine_era(arabic_text),
            "travel_history": self._extract_travel_history(arabic_text),
            "did_travel_for_hadith": self._check_travel_for_hadith(arabic_text),
            "memory_changes": self._extract_memory_changes(arabic_text),
            "known_tadlis": self._check_tadlis(arabic_text),
            "scholarly_reliability": self._extract_reliability_terms(arabic_text),
            "scholarly_evaluations": self._extract_evaluations(arabic_text),
            "teachers": self._extract_teachers(arabic_text),
            "students": self._extract_students(arabic_text),
            "death_details": self._extract_death_details(arabic_text),
            "additional_notes": self._extract_additional_notes(arabic_text)
        }

        return metadata

    def _extract_full_name(self, text: str) -> str:
        """Extract the full name from the beginning of the text."""
        # Names typically appear at the start before الإمام or after pattern
        lines = text.split('،')
        if lines:
            # Clean up the first part which usually contains the full name
            name = lines[0].strip()
            # Remove titles like الإمام، الحافظ, etc.
            name = re.sub(r'الإمام|الحافظ|القدوة|الشيخ|العلامة', '', name).strip()
            return name
        return ""

    def _extract_aliases(self, text: str) -> List[str]:
        """Extract aliases and kunyas (Abu, Ibn, etc.)."""
        aliases = []

        # Extract kunya (أبو ...)
        kunya_matches = re.findall(r'أبو\s+\w+(?:\s+\w+)?', text)
        aliases.extend(kunya_matches)

        # Extract nisba (البصري، التيمي، etc.)
        nisba_matches = re.findall(r'ال\w+ي(?=\s|،|\.)', text)
        aliases.extend(nisba_matches)

        return list(set(aliases))  # Remove duplicates

    def _extract_birth_year(self, text: str) -> Optional[int]:
        """Extract birth year in Hijri."""
        # Pattern: ولد سنة ... ومائة
        patterns = [
            r'ولد سنة\s+(\w+)\s+ومائة',
            r'مولده سنة\s+(\w+)\s+ومائة',
        ]

        for pattern in patterns:
            match = re.search(pattern, text)
            if match:
                year_word = match.group(1)
                year = self._arabic_number_to_int(year_word)
                if year:
                    return 100 + year
        return None

    def _extract_death_year(self, text: str) -> Optional[int]:
        """Extract death year in Hijri."""
        # Pattern: مات سنة ... ومائة OR توفي سنة ...
        patterns = [
            r'مات.*?سنة\s+(\w+)\s+وثمانين ومائة',
            r'مات.*?سنة\s+(\w+)\s+وسبعين ومائة',
            r'مات.*?سنة\s+(\w+)\s+وستين ومائة',
            r'توفي.*?سنة\s+(\w+)\s+ومائة',
            r'مات.*?سنة\s+(\d+)',
        ]

        for pattern in patterns:
            match = re.search(pattern, text)
            if match:
                year_text = match.group(1)
                # Check if it's already a number
                if year_text.isdigit():
                    return int(year_text)

                year = self._arabic_number_to_int(year_text)
                if year:
                    # Determine the century based on context
                    if 'ثمانين' in match.group(0):
                        return 180 + year
                    elif 'سبعين' in match.group(0):
                        return 170 + year
                    else:
                        return 100 + year
        return None

    def _extract_birthplace(self, text: str) -> str:
        """Extract birthplace."""
        # Usually indicated by nisba like البصري، المدني، الكوفي
        locations = {
            'البصري': 'Basra',
            'المدني': 'Medina',
            'الكوفي': 'Kufa',
            'البغدادي': 'Baghdad',
            'المصري': 'Egypt',
            'الشامي': 'Syria',
            'المكي': 'Mecca'
        }

        for arabic, english in locations.items():
            if arabic in text[:200]:  # Check in first part of text
                return english
        return ""

    def _extract_locations(self, text: str) -> List[str]:
        """Extract primary locations associated with the narrator."""
        locations = []
        location_keywords = {
            'البصرة': 'Basra',
            'المدينة': 'Medina',
            'الكوفة': 'Kufa',
            'بغداد': 'Baghdad',
            'مصر': 'Egypt',
            'الشام': 'Syria',
            'مكة': 'Mecca'
        }

        for arabic, english in location_keywords.items():
            if arabic in text:
                locations.append(english)

        return list(set(locations))

    def _determine_era(self, text: str) -> str:
        """Determine the era/generation of the narrator."""
        if 'الصحابة' in text or 'صحابي' in text:
            return "Companion (Sahabi)"
        elif 'التابعين' in text or 'تابعي' in text:
            return "Tabi'in (Successor)"
        elif 'أتباع التابعين' in text:
            return "Tabi' al-Tabi'in (Successors of Successors)"
        else:
            # Determine by death year
            death_year = self._extract_death_year(text)
            if death_year:
                if death_year < 110:
                    return "Tabi'in (Successor)"
                elif death_year < 200:
                    return "Tabi' al-Tabi'in (Successors of Successors)"
                elif death_year < 250:
                    return "Third Generation"
                else:
                    return "Later Generations"
        return "Unknown"

    def _extract_travel_history(self, text: str) -> List[str]:
        """Extract travel history for seeking knowledge."""
        travels = []
        # Look for رحل إلى، سافر إلى، etc.
        travel_patterns = [
            r'رحل إلى\s+(\w+)',
            r'سافر إلى\s+(\w+)',
        ]

        for pattern in travel_patterns:
            matches = re.findall(pattern, text)
            travels.extend(matches)

        return travels if travels else self._extract_locations(text)

    def _check_travel_for_hadith(self, text: str) -> bool:
        """Check if the narrator traveled for hadith."""
        travel_indicators = ['رحل', 'سافر', 'خرج في طلب', 'طلب العلم']
        return any(indicator in text for indicator in travel_indicators)

    def _extract_memory_changes(self, text: str) -> Dict:
        """Extract information about memory changes (اختلاط)."""
        memory_keywords = ['اختلط', 'تغير حفظه', 'ساء حفظه']

        for keyword in memory_keywords:
            if keyword in text:
                return {
                    "occurred": True,
                    "details": f"Memory issues mentioned: {keyword}"
                }

        return {"occurred": False, "details": None}

    def _check_tadlis(self, text: str) -> bool:
        """Check if narrator is known for tadlis (concealment)."""
        tadlis_keywords = ['مدلس', 'يدلس', 'التدليس']
        return any(keyword in text for keyword in tadlis_keywords)

    def _extract_reliability_terms(self, text: str) -> List[str]:
        """Extract reliability terminology used by scholars."""
        reliability_terms = []

        terms = {
            'ثقة': 'Thiqa (Trustworthy)',
            'صدوق': 'Saduq (Truthful)',
            'حافظ': 'Hafiz (Preserver)',
            'ضعيف': 'Da\'if (Weak)',
            'متروك': 'Matruk (Abandoned)',
            'كذاب': 'Kadhdhab (Liar)',
            'ثبت': 'Thabt (Firm/Reliable)'
        }

        for arabic, english in terms.items():
            if arabic in text:
                reliability_terms.append(english)

        return list(set(reliability_terms))

    def _extract_evaluations(self, text: str) -> List[Dict]:
        """Extract scholarly evaluations."""
        evaluations = []

        # Pattern: قال [scholar]: [evaluation]
        eval_pattern = r'قال\s+([^:：]+?)\s*[:：]\s*([^.。]+)'
        matches = re.findall(eval_pattern, text)

        for evaluator, evaluation in matches:
            evaluations.append({
                "evaluated_by": evaluator.strip(),
                "evaluation": evaluation.strip(),
                "justification": None,
                "year": None
            })

        return evaluations

    def _extract_teachers(self, text: str) -> List[str]:
        """Extract the narrator's teachers (شيوخ)."""
        teachers = []

        # Look for section starting with حدث عن or روى عن
        teacher_section = re.search(r'(?:حدث عن|روى عن)\s*[:：]\s*([^.]+)', text)

        if teacher_section:
            teacher_text = teacher_section.group(1)
            # Split by commas and 'و'
            teacher_list = re.split(r'[،,]|\s+و\s+', teacher_text)
            teachers = [t.strip() for t in teacher_list if t.strip()]

        return teachers

    def _extract_students(self, text: str) -> List[str]:
        """Extract the narrator's students (تلاميذ)."""
        students = []

        # Look for section starting with حدث عنه or روى عنه
        student_section = re.search(r'(?:حدث عنه|روى عنه)\s*[:：]\s*([^.]+)', text)

        if student_section:
            student_text = student_section.group(1)
            # Split by commas and 'و'
            student_list = re.split(r'[،,]|\s+و\s+', student_text)
            students = [s.strip() for s in student_list if s.strip()]

        return students

    def _extract_death_details(self, text: str) -> Optional[Dict]:
        """Extract detailed information about death."""
        death_info = {
            "year": self._extract_death_year(text),
            "month": None,
            "age": None,
            "location": None,
            "context": None
        }

        # Extract month
        months = ['المحرم', 'صفر', 'ربيع', 'جمادى', 'رجب', 'شعبان', 'رمضان', 'شوال', 'ذو القعدة', 'ذو الحجة']
        for month in months:
            if month in text:
                death_info["month"] = month
                break

        # Extract age
        age_match = re.search(r'وهو ابن\s+(\w+)\s+(?:و)?(\w+)?\s*سنة', text)
        if age_match:
            tens = self._arabic_number_to_int(age_match.group(1))
            ones = self._arabic_number_to_int(age_match.group(2)) if age_match.group(2) else 0
            if tens and ones is not None:
                death_info["age"] = tens + ones

        # Extract location
        for location in self._extract_locations(text):
            if f'مات ب{location}' in text or f'توفي ب{location}' in text:
                death_info["location"] = location
                break

        return death_info if death_info["year"] else None

    def _extract_additional_notes(self, text: str) -> List[str]:
        """Extract additional noteworthy information."""
        notes = []

        # Look for special mentions
        if 'موالي' in text or 'مولى' in text:
            mawla_match = re.search(r'مول[ىي]\s+([^،.]+)', text)
            if mawla_match:
                notes.append(f"Mawla (client) of {mawla_match.group(1).strip()}")

        if 'نزل في' in text:
            settlement_match = re.search(r'نزل في\s+([^،.]+)', text)
            if settlement_match:
                notes.append(f"Settled among {settlement_match.group(1).strip()}")

        if 'من كبار العلماء' in text:
            notes.append("Considered among the major scholars")

        return notes

    def _arabic_number_to_int(self, word: str) -> Optional[int]:
        """Convert Arabic number words to integers."""
        numbers = {
            'واحد': 1, 'اثنين': 2, 'ثلاث': 3, 'أربع': 4, 'خمس': 5,
            'ست': 6, 'سبع': 7, 'ثمان': 8, 'تسع': 9, 'عشر': 10,
            'إحدى': 1, 'اثنتين': 2, 'ثلاثة': 3, 'أربعة': 4, 'خمسة': 5,
            'ستة': 6, 'سبعة': 7, 'ثمانية': 8, 'تسعة': 9
        }

        word = word.strip()
        return numbers.get(word)

    def export_to_json(self, metadata: Dict, filename: str = "narrator_metadata.json"):
        """Export metadata to JSON file."""
        output = {"narrators": [metadata]}
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(output, f, ensure_ascii=False, indent=2)
        print(f"Metadata exported to {filename}")


def process_dataframe(df, bio_column='bio', narrator_id_column=None):
    """
    Process a DataFrame with biographical texts and extract metadata for all rows.

    Args:
        df: pandas DataFrame containing biographical texts
        bio_column: Name of the column containing Arabic biographical texts
        narrator_id_column: Optional column name for narrator IDs. If None, will auto-generate.

    Returns:
        pandas DataFrame with extracted metadata as new columns
    """
    import pandas as pd
    from tqdm import tqdm

    extractor = NarratorBiographyExtractor()

    # Store all extracted metadata
    all_metadata = []

    # Process each row with progress bar
    for idx, row in tqdm(df.iterrows(), total=len(df), desc="Extracting narrator metadata"):
        bio_text = row[bio_column]

        # Generate narrator ID
        if narrator_id_column and narrator_id_column in df.columns:
            narrator_id = row[narrator_id_column]
        else:
            narrator_id = f"N{idx+1:04d}"  # N0001, N0002, etc.

        try:
            # Extract metadata
            metadata = extractor.extract_narrator_metadata(bio_text, narrator_id)
            all_metadata.append(metadata)
        except Exception as e:
            print(f"Error processing row {idx}: {e}")
            # Append empty metadata on error
            all_metadata.append({"narrator_id": narrator_id, "error": str(e)})

    # Convert to DataFrame
    metadata_df = pd.DataFrame(all_metadata)

    return metadata_df


def process_dataframe_expanded(df, bio_column='bio', narrator_id_column=None):
    """
    Process DataFrame and expand nested fields into separate columns.
    Returns a wide-format DataFrame with all extracted information.
    """
    import pandas as pd
    from tqdm import tqdm

    extractor = NarratorBiographyExtractor()

    rows_data = []

    for idx, row in tqdm(df.iterrows(), total=len(df), desc="Extracting metadata"):
        bio_text = row[bio_column]

        if narrator_id_column and narrator_id_column in df.columns:
            narrator_id = row[narrator_id_column]
        else:
            narrator_id = f"N{idx+1:04d}"

        try:
            metadata = extractor.extract_narrator_metadata(bio_text, narrator_id)

            # Flatten the metadata
            flat_data = {
                'narrator_id': metadata['narrator_id'],
                'full_name': metadata['full_name'],
                'aliases': ', '.join(metadata['aliases']) if metadata['aliases'] else '',
                'birth_year': metadata['birth_year'],
                'death_year': metadata['death_year'],
                'birthplace': metadata['birthplace'],
                'primary_locations': ', '.join(metadata['primary_locations']) if metadata['primary_locations'] else '',
                'era': metadata['era'],
                'did_travel_for_hadith': metadata['did_travel_for_hadith'],
                'memory_changes_occurred': metadata['memory_changes']['occurred'],
                'memory_changes_details': metadata['memory_changes']['details'],
                'known_tadlis': metadata['known_tadlis'],
                'scholarly_reliability': ', '.join(metadata['scholarly_reliability']) if metadata['scholarly_reliability'] else '',
                'num_teachers': len(metadata['teachers']),
                'num_students': len(metadata['students']),
                'teachers_list': ' | '.join(metadata['teachers']) if metadata['teachers'] else '',
                'students_list': ' | '.join(metadata['students']) if metadata['students'] else '',
                'num_evaluations': len(metadata['scholarly_evaluations']),
                'death_age': metadata['death_details']['age'] if metadata['death_details'] else None,
                'death_location': metadata['death_details']['location'] if metadata['death_details'] else None,
                'additional_notes': ' | '.join(metadata['additional_notes']) if metadata['additional_notes'] else ''
            }

            # Add original columns from df
            for col in df.columns:
                if col not in flat_data:
                    flat_data[f'original_{col}'] = row[col]

            rows_data.append(flat_data)

        except Exception as e:
            print(f"Error processing row {idx}: {e}")
            rows_data.append({'narrator_id': narrator_id, 'error': str(e)})

    return pd.DataFrame(rows_data)


```

```python
# Example 2: DataFrame processing
# Process with nested structure
print("\n--- Processing with nested metadata ---")
result_nested = process_dataframe(df, bio_column='bio')
print("\nResult shape:", result_nested.shape)
print("\nColumns:", result_nested.columns.tolist())
print("\nFirst row full_name:", result_nested.iloc[0]['full_name'])

# Process with expanded columns
print("\n--- Processing with expanded columns ---")
result_expanded = process_dataframe_expanded(df, bio_column='bio')
print("\nResult shape:", result_expanded.shape)
print("\nColumns:", result_expanded.columns.tolist())
print("\nSample data:")
print(result_expanded[['narrator_id', 'full_name', 'birth_year', 'death_year', 'era']].head())

# Export results
print("\n--- Exporting results ---")
result_expanded.to_csv('narrator_metadata_expanded.csv', index=False, encoding='utf-8-sig')
result_nested.to_json('narrator_metadata_nested.json', orient='records', force_ascii=False, indent=2)

print("\nFiles saved:")
print("- narrator_metadata_expanded.csv (wide format)")
print("- narrator_metadata_nested.json (nested format)")

print("\n" + "=" * 60)
print("Processing Complete!")
print("=" * 60)
```
